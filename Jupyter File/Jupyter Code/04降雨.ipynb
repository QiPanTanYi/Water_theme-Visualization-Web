```python
import pandas as pd
import json
```

```python
df=pd.read_excel("../data/2021年各月份各省份主要城市的降雨情况.xlsx")
df
```

```python
data={}
for i in range(33):
    df1=list(df.iloc[i,:])
    data[df1[0]]=df1[1:]
data
```

```python
with open("../json/降雨.json","w",encoding="utf-8") as f:
    f.write(json.dumps(data,ensure_ascii=False))
```
